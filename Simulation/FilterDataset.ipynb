```python
def filter_according(df,long='pickup_longitude',lat = 'pickup_latitude',
                     min_longitude = -74.02,
                        max_longitude = -73.7,
                        min_latitude = 40.7,
                        max_latitude = 40.85,
                          ):
    # Create boolean masks for filtering
    longitude_mask = (df[long] >= min_longitude) & (df[long] <= max_longitude)
    latitude_mask = (df[lat] >= min_latitude) & (df[lat] <= max_latitude)
    # Apply both masks to get the final subset
    filtered_df = df[longitude_mask & latitude_mask]
    filtered_df[long] = np.abs(filtered_df[long])
    return filtered_df
```

```python
min_longitude = -74.02
max_longitude = -73.7
min_latitude = 40.7
max_latitude = 40.85

filtered_df = filter_according(df,long='pickup_longitude',lat = 'pickup_latitude',
                        min_longitude = -74.02,
                        max_longitude = -73.7,
                        min_latitude = 40.7,
                        max_latitude = 40.85)
```
